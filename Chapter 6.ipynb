# Reindexing

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
```

```python
obj
```

```python
 obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e'])  #if any index is not there then NaN
obj2 
```

```python
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4])
obj3
```

```python
obj3.reindex(range(6))
```

```python
obj3.reindex(range(6), method='ffill')  #forward fill if any value is NaN after blue then forward the blue value and so on 
```

```python
obj3.reindex(range(6), method='bfill')
```

# Reindexing on Frame

```python
frame = pd.DataFrame(np.arange(9).reshape((3, 3)),
  index=['a', 'c', 'd'],
  columns=['Ohio', 'Texas', 'California'])
frame
```

```python
frame2 = frame.reindex(['a', 'b', 'c', 'd'])
frame2
```
